```
trainfile = r'/gdrive/Projects/House price prediction/train.csv'
trainData = pd.read_csv(trainfile)  #creates a dataframe
testfile = r'/gdrive/Projects/House price prediction/test.csv'
testData = pd.read_csv(testfile)  #creates a dataframe
print(trainData.shape)
print(testData.shape)
```

```
#Extract Target Column before doing missing value substitutions and one-hot encoding======
Target_Train_Cols = trainData["SalePrice"]#make copy of target column
trainData = trainData.drop(["SalePrice"], axis=1) #extracting training data without the target column

print(trainData.shape)
```

```
#Combine Train data and test data first so that the SAME COLUMNS are DROPPED in each
combined_Data = pd.concat([trainData, testData], keys=[0,1])

```

```
# Allows us to see the number of missing values in a column. 
def missing_values_table(df):
        mis_val = df.isnull().sum()
        mis_val_dtypes = df.dtypes
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent,mis_val_dtypes], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values', 2:'dtypes'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns

missing_values_table(combined_Data)
```

```
#DROP COLUMNS WITH LOTS OF MISSING VALUES===============================

del combined_Data['PoolQC']
del combined_Data['MiscFeature']
del combined_Data['Alley']
del combined_Data['Fence']
del combined_Data['FireplaceQu']
```

```
# Allows us to see the number of missing values in a column. 
def missing_values_table(df):
        mis_val = df.isnull().sum()
        mis_val_dtypes = df.dtypes
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent,mis_val_dtypes], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values', 2:'dtypes'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns

missing_values_table(combined_Data)
```

```
#NOW IMPUTE MISSING VALUES FOR THE OTHER COLUMNS=========================
#IMPUTE (SUBSTITUTE) MEAN VALUES FOR NaN IN NUMERIC COLUMNS 
numeric=combined_Data.select_dtypes(include=['int','float64']).columns
for num in numeric:
  combined_Data[num]=combined_Data[num].fillna(combined_Data[num].mean())

#IMPUTE (SUBSTITUTE) MODE VALUES FOR NaN IN CATEGORICAL COLUMNS
combined_cat_cols = combined_Data.select_dtypes(exclude=['int','float64']).columns#selecting the categorical columns
for colss in combined_cat_cols:
  if(combined_Data.iloc[0][colss]=="N"):
        combined_Data[colss]=combined_Data[colss].fillna("N")
  else:
    combined_Data[colss]=combined_Data[colss].fillna(combined_Data[colss].mode())
  
print(trainData.head(20))
```

```
#DO ONE-HOT ENCODING ON CATEGORICAL VARIABLES==============================================
#The below function returns a list of categorical features which are not numeric. 
combined_cat_cols = combined_Data.select_dtypes(exclude=['float','int']).columns #selecting the categorical columns
print(combined_cat_cols.shape)
print(combined_cat_cols)

#If there are categorical columns which are encoded as numeric ones 
#then we need to explicitly enter the column names in a list and concatenate the two lists in python.
#ONE-HOT ENCODING-generate one-hot encoding on a common basis -THIS TAKES 30 MINS

combined_Data = pd.get_dummies(combined_Data,combined_cat_cols)
combined_Data.head(10)
```

```
#Separate Train data and test data
trainData = combined_Data.xs(0)
testData = combined_Data.xs(1)
print(trainData.shape)
print(testData.shape)

trainData=pd.concat([trainData,Target_Train_Cols], axis=1)
print(trainData.shape)
```
